```python
MID_QUARTER_FEEDBACK_IDS = [
    105963657, 105981784, 106069785, 405399441, 405747025, 405753059,
    405914788, 405960662, 505917606, 505917696, 505984101, 605721755,
    705646261, 705826846, 805916243, 805956658, 905698428, 905985636, 906178455,
]

END_QUARTER_FEEDBACK_IDS = [
    105963657, 105981784, 106069785, 206179477, 405399441, 405747025,
    405753059, 405914788, 406225690, 505917606, 505984101, 705646261,
    705826846, 805916243, 805956658, 806024406, 905698428, 906046988, 906178455,
]


def print_feedback_students(grades):
    # Get names of students who filled out the mid-quarter feedback
    mid_quarter_students = grades.loc[
        grades['SIS User ID'].isin(MID_QUARTER_FEEDBACK_IDS), 'Student'
    ]
    print("Students who filled out the mid-quarter feedback:")
    print(mid_quarter_students.tolist())

    # Get names of students who filled out the final-quarter feedback
    end_quarter_students = grades.loc[
        grades['SIS User ID'].isin(END_QUARTER_FEEDBACK_IDS), 'Student'
    ]
    print("\nStudents who filled out the final-quarter feedback:")
    print(end_quarter_students.tolist())
    
    
print_feedback_students(grades)
```
